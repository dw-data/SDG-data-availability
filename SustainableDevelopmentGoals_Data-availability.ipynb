# How is the data availability for the Sustainable Development Goal indicators?

```python
df = pd.read_csv("SDG-data-availability.csv", delimiter=";")
```

```python
df
```

```python
number_sdg_indicators = df['TOTAL'].sum()
number_regular_data = df['Regular'].sum()
number_irregular_data = df['Irregular'].sum()
number_missing_data = df['Missing'].sum()
number_uncategorized_data = df['Uncategorised'].sum()
print("There are", number_sdg_indicators,"SDG indicators. Of these")
print(number_regular_data, "have data available regularly. This equals",int(round(number_regular_data/number_sdg_indicators*100)),"percent of all SDG indicators.")
print(number_irregular_data, "have data available irregularly. This equals",int(round(number_irregular_data/number_sdg_indicators*100)),"percent of all SDG indicators.")
print(number_missing_data, "have data no data available. This equals",int(round(number_missing_data/number_sdg_indicators*100)),"percent of all SDG indicators.")
print(number_uncategorized_data, "have no details on data availability. This equals",int(round(number_uncategorized_data/number_sdg_indicators*100)),"percent of all SDG indicators.")
```
